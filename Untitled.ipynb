```python

import matplotlib.pyplot as plt


method =input("Enetr the Algorithm:")


initial_position=int(input("Enetr the current position:"))

disk_requests=list(map(int,input("Enter the request positions:\n").split()))

total_head_movement=0
if method.upper()=='FCFS':
    # FCFS order is the order of the disk_requests list
    fcfs_order = disk_requests

    # Disk movement path including the initial position
    disk_path = [initial_position] + fcfs_order

    request_sequence = list(range(len(disk_path)))

    # Plotting the disk movement
    plt.figure(figsize=(10, 8))
    plt.plot(request_sequence,disk_path, marker='o', linestyle='-', color='b')
    plt.xticks(request_sequence,)
    plt.yticks(sorted(disk_path))
    plt.xlabel('Request Sequence')
    plt.ylabel('Cylinder Number')
    plt.title('FCFS Disk Scheduling Disk Movement')
    plt.show()
elif method.upper()=='SSTF':
    requests = disk_requests.copy()
    current_position = initial_position
    sequence = [current_position]
    
    while requests:
        # Find the closest request
        closest_request = min(requests, key=lambda x: abs(x - current_position))
        sequence.append(closest_request)
        requests.remove(closest_request)
        current_position = closest_request
    # Get the SSTF sequence
    sstf_sequence = sequence
    
    #Request sequence for x-axis
    request_sequence = list(range(len(sstf_sequence)))

    # Plotting the disk movement
    plt.figure(figsize=(10, 6))
    plt.plot(request_sequence, sstf_sequence, marker='o', linestyle='-', color='black')
    plt.xticks(request_sequence)
    plt.yticks(sorted(set(sstf_sequence)))
    plt.xlabel('Request Sequence')
    plt.ylabel('Cylinder Number')
    plt.title('SSTF Disk Scheduling Disk Movement')
    plt.show()
elif method.upper()=='SCAN':
    direction='up'
    requests = sorted(disk_requests)
    sequence = [initial_position]
    
    if direction == 'up':
        upward_requests = [req for req in requests if req >= initial_position]
        downward_requests = [req for req in requests if req < initial_position]
        
        sequence.extend(upward_requests)
        sequence.append(int(input("Enter the maximum disk:")))
        if downward_requests:
            sequence.extend(reversed(downward_requests))
    else:
        downward_requests = [req for req in requests if req <= initial_position]
        upward_requests = [req for req in requests if req > initial_position]
        
        sequence.extend(reversed(downward_requests))
        if upward_requests:
            sequence.append(0)
            sequence.extend(upward_requests)
    
    

    scan_sequence = sequence

    # Request sequence for x-axis
    request_sequence = list(range(len(scan_sequence)))

    # Plotting the disk movement
    plt.figure(figsize=(10, 6))
    plt.plot(request_sequence, scan_sequence, marker='o', linestyle='-', color='black')
    plt.xticks(request_sequence)
    plt.yticks(sorted(set(scan_sequence)))
    plt.xlabel('Request Sequence')
    plt.ylabel('Cylinder Number')
    plt.title('SCAN Disk Scheduling Disk Movement')
    plt.show()
elif method.upper()=="CSCAN":
    requests = sorted(disk_requests)
    sequence = [initial_position]
    
    upward_requests = [req for req in requests if req >= initial_position]
    downward_requests = [req for req in requests if req < initial_position]
    
    sequence.extend(upward_requests)
    sequence.append(int(input("Enter the maximum disk:")))

    if downward_requests:
        sequence.extend(downward_requests)
    
    

    # Get the C-SCAN sequence
    c_scan_sequence =sequence

    # Request sequence for x-axis
    request_sequence = list(range(len(c_scan_sequence)))

    # Plotting the disk movement
    plt.figure(figsize=(10, 6))
    plt.plot(request_sequence, c_scan_sequence, marker='o', linestyle='-', color='black')
    plt.xticks(request_sequence)
    plt.yticks(sorted(set(c_scan_sequence)))
    plt.xlabel('Request Sequence')
    plt.ylabel('Cylinder Number')
    plt.title('C-SCAN Disk Scheduling Disk Movement')
    plt.show()
elif method.upper()=="LOOK":
    requests = sorted(disk_requests)
    sequence = [initial_position]
    direction='up'
    
    if direction == 'up':
        upward_requests = [req for req in requests if req >= initial_position]
        downward_requests = [req for req in requests if req < initial_position]
        
        sequence.extend(upward_requests)
        if downward_requests:
            sequence.extend(reversed(downward_requests))
    else:
        downward_requests = [req for req in requests if req <= initial_position]
        upward_requests = [req for req in requests if req > initial_position]
        
        sequence.extend(reversed(downward_requests))
        if upward_requests:
            sequence.extend(upward_requests)
    
    
    look_sequence = sequence

    # Request sequence for x-axis
    request_sequence = list(range(len(look_sequence)))

    # Plotting the disk movement
    plt.figure(figsize=(10, 6))
    plt.plot(request_sequence, look_sequence, marker='o', linestyle='-', color='black')
    plt.xticks(request_sequence)
    plt.yticks(sorted(set(look_sequence)))
    plt.xlabel('Request Sequence')
    plt.ylabel('Cylinder Number')
    plt.title('LOOK Disk Scheduling Disk Movement')
    plt.show()
elif method.upper()=='CLOOK':
    requests = sorted(disk_requests)
    sequence = [initial_position]
    
    # Separate requests into those greater than and less than the initial position
    upward_requests = [req for req in requests if req >= initial_position]
    downward_requests = [req for req in requests if req < initial_position]
    
    # Serve upward requests first
    sequence.extend(upward_requests)
    
    # If there are downward requests, jump to the beginning and serve them
    if downward_requests:
        sequence.extend(downward_requests)
    

    # Get the C-LOOK sequence
    c_look_sequence = sequence

    # Request sequence for x-axis
    request_sequence = list(range(len(c_look_sequence)))

    # Plotting the disk movement
    plt.figure(figsize=(10, 6))
    plt.plot(request_sequence, c_look_sequence, marker='o', linestyle='-', color='black')
    plt.xticks(request_sequence)
    plt.yticks(sorted(set(c_look_sequence)))
    plt.xlabel('Request Sequence')
    plt.ylabel('Cylinder Number')
    plt.title('C-LOOK Disk Scheduling Disk Movement')
    plt.show()
```
